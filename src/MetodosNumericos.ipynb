```python
data = pd.read_csv('data/datos.csv', index_col=0)
data.head()
```

```python
X = data[['math','female','socst','read']]
Y = data.science
lr = LinearRegression()
lr.fit(X,Y)
```

```python
print(f'hat_b0 = {lr.intercept_}')
print(f'hat_b1 = {lr.coef_[0]}')
print(f'hat_b2 = {lr.coef_[1]}')
print(f'hat_b3 = {lr.coef_[2]}')
print(f'hat_b4 = {lr.coef_[3]}')
```

```python
interc = np.ones([len(data['female']),1])
```

```python
data['intercepto'] = interc
datos_prueba_X = data.drop('science', axis=1)
A = datos_prueba_X.T.dot(datos_prueba_X)
b = datos_prueba_X.T.dot(data['science'])
A
```

```python
coefs = mn.backward_substi(mn.Gauss(A.to_numpy(),b.to_numpy())[0], mn.Gauss(A.to_numpy(),b.to_numpy())[1])
nombres_variables = datos_prueba_X.columns.tolist()

# Define la cadena de formato
formato = "{:<12} {:>10.4f} "
print("=" * 23) 
print("{:<12} {:>10} ".format("Variable", "Coef"))
print("-" * 23) 

for nombre, coeficiente in zip(nombres_variables, coefs):
    print(formato.format(nombre, coeficiente, 0, 0))
```

```python
formula = 'science ~ math + female + socst + read'
model = sm.ols(formula = formula, data = data)
fitted = model.fit()
print(fitted.summary())
```

# QR method

```python
A = np.array([[12, -51, 4],
    		[6, 167, -68],
    		[-4, 24, -41]],float)
```

```python
b = np.array([3, 5, -12],float)

n = A.shape
Q, R = mn.metodoGramm(A,b,n[0])
```

```python
Qt = np.transpose(Q)

print("\nPara calcular la solucion 'x' del problema, resolveremos R * x = Qt(Q transpuesta) * b",end="\n")

print('\nMatriz Q transpuesta:\n', Qt.round(7))

print('\nVector b:\n', b.round(7))
```
